## 1.导入模型和分词器，读取数据集，制作词典 Import models and tokenizers, read data sets, and make dictionaries

```python
import copy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import shap
import scipy as sp
import torch
import os
os.chdir('/content/drive/MyDrive/Datasets')
# from datasets import load_dataset


data = pd.read_csv('headline-sarcasm-shap.tsv', delimiter="\t")
print(data.head())

tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-irony", use_fast=True)
model = AutoModelForSequenceClassification.from_pretrained("cardiffnlp/twitter-roberta-base-irony").cuda()

id2label = model.config.id2label
label2id = model.config.label2id
labels = sorted(label2id, key=label2id.get)
```

## 2.显示不同标签的样本的分布情况 Distribution of sarcasm labels


```python
ax = data.is_sarcastic.map(id2label).value_counts().plot.bar()
```

##3.计算shap所需要的函数 Functions needed to calculate shap

```python
def f(x):
    tv = torch.tensor([tokenizer.encode(v, padding='max_length', max_length=128,truncation=True) for v in x]).cuda()
    attention_mask = (tv!=0).type(torch.int64).cuda()
    outputs = model(tv,attention_mask=attention_mask)[0].detach().cpu().numpy() #
    scores = (np.exp(outputs).T / np.exp(outputs).sum(-1)).T
    val = sp.special.logit(scores)
    return val
```

## 4.用explanier计算样本的各个词语的shap值 Use explanier to calculate the shap value of each word in the sample



```python
explainer = shap.Explainer(f,tokenizer,output_names=labels)
shap_values = explainer(data['headline'][:])
```

删除部分的主函数，实现删除功能 Deletepart main function

```python
# 删除句子中与标签相关性最高的两个词 Delete the two words that are most relevant to the tag in the sentence
def delete(sentence):
  shap_values = explainer(sentence['headline'])
  arrays = shap_values.values
  datas = shap_values.data
  print('The origin sentence is: ', sentence['headline'].values)
  try: #
    if sentence['is_sarcastic'].values == 0:
      arrays0 = arrays[0, 1:][:, 0]
      index0_1 = arrays0.argmax()
      index0_2 = np.argsort(arrays0)[-2]
      replace_data_1 = datas[0, 1:][index0_1]
      replace_data_2 = datas[0, 1:][index0_2]
      sentence['headline'] = sentence['headline'].str.replace(replace_data_1, "")
        # print(sentence['headline'])
      print("After deleting the first related word: ", sentence['headline'].values, "The related words deleted is", replace_data_1)
      sentence['headline'] = sentence['headline'].str.replace(replace_data_2, "")
        # print(sentence['headline'])
      print("After deleting the second related word: ", sentence['headline'].values, "The related words deleted is", replace_data_2)
      #series = sentence['headline'] 原来的
      data = sentence['headline'].values
      series = pd.Series(data.tolist())
    else:
      arrays1 = arrays[0, 1:][:, 1]
      index1_1 = arrays1.argmax()
      index1_2 = np.argsort(arrays1)[-2]
      replace_data_1 = datas[0, 1:][index1_1]
      replace_data_2 = datas[0, 1:][index1_2]
      sentence['headline'] = sentence['headline'].str.replace(replace_data_1, "")
      print("After deleting the first related word: ", sentence['headline'].values, "The related words deleted is", replace_data_1)
      # print(sentence['headline'])
      sentence['headline'] = sentence['headline'].str.replace(replace_data_2, "")
      print("After deleting the second related word: ", sentence['headline'].values, "The related words deleted is", replace_data_2)
      # print(sentence['headline'])
      # series = sentence['headline'] 原来的
      data = sentence['headline'].values
      series = pd.Series(data.tolist())
  except:
      series = sentence['headline']
      print('The sentence is too long')
  return series
```

循环删除dataframe中，与是否讽刺最相关的两个词 Cyclically delete the two words most relevant to whether or not scarcastic is in the dataframe

```python
deleted = pd.DataFrame(columns=('deleted_sentence_x','Label','no_deleted_sentence_y'))
for i in range(0, len(input_df)):
    # print(input_df.iloc[i]['is_sarcastic'])
    # print(input_df.iloc[i]['headline'])
    temp_dict = {'headline': [input_df.iloc[i]['headline']], "is_sarcastic": [input_df.iloc[i]['is_sarcastic']]}
    temp_df = pd.DataFrame(temp_dict)
    # print(temp_df)
    series = delete(temp_df)
    deleted_sentence_x = series.values
    Label = 'Label_1'
    no_deleted_sentence_y = input_df.iloc[i]['headline']
    # deleted = deleted.append((deleted_sentence_x, Label, no_deleted_sentence_y), ignore_index=True)
    deleted = deleted.append(pd.DataFrame({'deleted_sentence_x':[deleted_sentence_x],'Label':[Label],'no_deleted_sentence_y':[no_deleted_sentence_y]}),ignore_index=True)
    print()
# deleted.to_csv('deleted.csv', sep="\t")
```
